```python
import pandas as pd
unrate = pd.read_csv('unrate.csv')
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
print(unrate.head(12))
```

```python
import matplotlib.pyplot as plt
plt.plot()  ## for plotting 
plt.show()  ## for displaying
%matplotlib inline      
##for displaying in the Jupyter notebook
```

### ADDING DATA

```python
import matplotlib.pyplot as plt
plt.plot(unrate['DATE'].loc[0:11],unrate['VALUE'].loc[0:11])
plt.show()
```

```python
plt.plot(unrate['DATE'].loc[0:11],unrate['VALUE'].loc[0:11])
plt.xticks(rotation=90) ## rotating the x axis by 90 degrees
plt.show() 
```

```python
plt.plot(unrate['DATE'].loc[0:11],unrate['VALUE'].loc[0:11])
plt.xticks(rotation=90)
plt.xlabel("Month")  ## adding label to the X axis
plt.ylabel("Unemployment Rate") ## adding label to the Y axis
plt.title("Monthly Unemployment Trends 1948") ## giving title to the graph
```
