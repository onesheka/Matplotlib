# Multiple Plots

```python
## done in the line charts
import pandas as pd
import matplotlib.pyplot as plt

unrate = pd.read_csv('unrate.csv')
unrate['DATE'] = pd.to_datetime(unrate['DATE'])

plt.plot(unrate['DATE'].loc[0:11],unrate['VALUE'].loc[0:11])
plt.xticks(rotation=90)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
plt.show()
```

## MATPLOTLIB CLASSES

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
```

## GRID POSITIONING

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(unrate['DATE'].loc[0:11],unrate['VALUE'].loc[0:11])
ax2.plot(unrate['DATE'].loc[12:23],unrate['VALUE'].loc[12:23])
plt.show()
```

## FORMATTING AND SPACING

```python
fig = plt.figure(figsize=(12,9))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(unrate['DATE'].loc[0:11],unrate['VALUE'].loc[0:11])
ax2.plot(unrate['DATE'].loc[12:23],unrate['VALUE'].loc[12:23])
ax1.set_title("Monthly Unemployment Trends, 1948") ## here it's set_title, not just title
ax2.set_title("Monthly Unemployment Trends, 1949")
plt.show()
```

```python
fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(5,1,1)
ax2 = fig.add_subplot(5,1,2)
ax3 = fig.add_subplot(5,1,3)
ax4 = fig.add_subplot(5,1,4)
ax5 = fig.add_subplot(5,1,5)

ax1.plot(unrate[0:12]['DATE'],unrate[0:12]['VALUE'])
ax2.plot(unrate[12:24]['DATE'],unrate[12:24]['VALUE'])
ax3.plot(unrate[24:36]['DATE'],unrate[24:36]['VALUE'])
ax4.plot(unrate[36:48]['DATE'],unrate[36:48]['VALUE'])
ax5.plot(unrate[48:60]['DATE'],unrate[48:60]['VALUE'])
## we can aslo use loop instead of typing for every indidual year
plt.show()
```

## OVERLAYING LINE CHARTS

```python
unrate['MONTH'] = unrate['DATE'].dt.month ## Creating a new column that contains all months using pandas dt.
plt.figure(figsize = (6,3))
plt.plot(unrate[0:12]['MONTH'],unrate[0:12]['VALUE'])
plt.plot(unrate[12:24]['MONTH'],unrate[12:24]['VALUE'])
plt.show()
```

```python
plt.figure(figsize = (10,6))
color = ['red','blue','green','orange','black']
for i in range(0,5):
    start = i*12
    end = (i+1)*12
    col = color[i]
    plt.plot(unrate[start:end]['MONTH'], unrate[start:end]['VALUE'],color = col)
    
plt.show()
```

## Creating legends

```python
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
lab  = 1948
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=lab+i)

plt.legend(loc="upper left")
plt.show()
```

## Final Multiple Plot chart

```python
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948 + i)
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=label)
plt.legend(loc='upper left')
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.xlabel("Month, Integer")
plt.ylabel("Unemployment Rate, Percent")
plt.show()
```
